```python
import numpy as np
import pandas as pd
import csv
```

```python
lenders = pd.read_csv("lenders.csv")
```

```python
name_gender_dict ={}
with open('wgnd_noctry.csv', mode='r') as infile:
    name_gender = csv.reader(infile)
    for rows in name_gender:
        name_gender_dict.update({rows[0]:rows[1]})
```

```python
gender = []
for name in lenders["DISPLAY_NAME"]:
    name_str = str(name).upper()
    if name_str in name_gender_dict:
        gender.append(name_gender_dict.get(name_str))
    else:
        gender.append(None)
```

```python
lenders["GENDER"]=gender
```

```python
lenders.to_csv("lenders_update.csv")
```
