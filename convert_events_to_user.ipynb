```python
import pandas as pd
import numpy as np
import random
import time

random.seed(1)

headers = pd.read_csv('headers_first_1000.csv',index_col = 0).columns

#extract column level data from november dataset
#option to remove bots, remove server actions, and remove non-account users
def extract_col(f, cols, 
                r_bot = True, 
                r_server = True, 
                rnon_user = True):
    tic = time.time()
    #choose only the columns that we need
    #browser type to get rid of bots
    #platform to get rid of server actions
    #user_id to get rid of users without accounts
    temp = pd.read_csv(f, names = headers, usecols = ['BR_TYPE','PLATFORM','USER_ID',*cols])
    toc = time.time()
    print('Time elapsed: ' + str(round(toc - tic, 2)) + 's')
    if r_bot:
        temp = temp[temp['BR_TYPE'] != 'Bot/Crawler']
    if r_server:
        temp = temp[temp['PLATFORM'] != 'srv']
    if rnon_user:
        temp = temp[~temp['USER_ID'].isnull()]
    tic = time.time()
    print('Narrowed data: ' + str(round(tic - toc, 2)) + 's')
    return temp

d = [extract_col('novdata/nov'+fi+'.csv', ['GEO_LATITUDE',
                                   'GEO_LONGITUDE',
                                   'GEO_COUNTRY', 
                                   'GEO_REGION']) for fi in ['1','2','3','4','5']]
fulld = pd.concat(d)
fulld
```

```python
fulld.drop(index=0,inplace  = True)
```

```python
uniques = fulld.groupby('USER_ID')
```

```python
unique_users = uniques.first()
unique_users['size'] = uniques.size()
unique_users.index = unique_users.index.astype(int)
```

```python
unique_users.to_csv('nov_event_users.csv')
```
